# Handle Categorical Features - 

## 3. Count Or Frequency Encoding

```python
import pandas as pd
df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
df.head()
```

```python
columns=[1,3,5,6,7,8,9,13]

train_set=df[columns]
```

```python
train_set
```

```python
train_set.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
train_set.head()
```

```python
for feature in train_set.columns[:]:
    print(feature,":",len(train_set[feature].unique()),'categories')
```

```python
train_set['Country'].value_counts()
```

```python
## to dict 
train_set['Country'].value_counts().to_dict()
```

```python
country_map=train_set['Country'].value_counts().to_dict()
```

```python
## mapping the country with the no of times that country appeared
train_set['Country']=train_set['Country'].map(country_map)
train_set.head(20)
```

### Advantages
    Easy To Use
    Not increasing feature space 

### Disadvantages
    It will provide same weight if the frequencies are same

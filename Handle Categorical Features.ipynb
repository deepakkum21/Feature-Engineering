## Onehotencoding with many categories in a feature

```python
df=pd.read_csv('mercedes.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
df=pd.read_csv('mercedes.csv',usecols=["X0","X1","X2","X3","X4","X5","X6"])
df.head()
```

```python
## for finding unique no of values in each cols
for i in df.columns:
    print(len(df[i].unique()))
```

### note:- 
#### when there are many categories for a feature then consider the top 10 most frequent categories and do onehot encoding for those 10
#### the remaining categories can be droped or ignored

```python
df['X0'].value_counts()
```

```python
df['X0'].value_counts().sort_values(ascending=False).head(10)
```

```python
df.X1.value_counts().sort_values(ascending=False).head(10).index
```

```python
lst_10=df.X1.value_counts().sort_values(ascending=False).head(10).index
lst_10=list(lst_10)
```

```python
lst_10
```

```python
## replacing the least frequent category with 0 and the most 10 frequet category with 1

import numpy as np
for categories in lst_10:
    df[categories]=np.where(df['X1']==categories,1,0)
```
